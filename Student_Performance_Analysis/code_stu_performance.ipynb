```python
# For range, we know that range is equal to difference of maximum and minimum value. so
subjects=["math score","reading score", "writing score" ]
for subject in subjects:
  a=df[subject].max()-df[subject].min()
  print("Range of",subject,"is",a)
```

```python
#For variance,use var() function
subjects=["math score","reading score", "writing score" ]
for subject in subjects:
  v=df[subject].var()
  print("Variance of",subject,"is",v)
```

```python
#For standard deviation,
subjects=["math score","reading score", "writing score" ]
for subject in subjects:
  s=df[subject].std()
  print("Standard deviation of",subject,"is",s)
```

Standard Deviation tells how spread out the data is from the average/mean .It tells variability of data. Since standard deviation of writing score is highest which is 15.19565,which tells that **Writing score is highly variable**.This shows that students have very different scores in writing.

```python
# for Boxplot
subjects=["math score","reading score", "writing score" ]
for subject in subjects:
  sns.boxplot(df[subject])
  plt.title(subject)
  plt.xlabel(subject)
  plt.ylabel('Frequency')
  plt.show()
```

Yes, there are very few outliers in data of all subjects . Maths have most outlier among all which indicates that  student find it challenging.Reading and Writing also have outliers but very few.

```python
# InterQuantile Range
subjects=["math score","reading score", "writing score" ]
for subject in subjects:
  q1=df[subject].quantile(0.25)
  q3=df[subject].quantile(0.75)
  iqr=q3-q1
  lower=q1-1.5*iqr
  upper=q3+1.5*iqr
  outliers=df[(df[subject]<lower) | (df[subject]>upper)]  # filter based on condition
  print("Outliers in",subject,":",len(outliers))
  print(outliers)
  print(" **********************************************************")
```

As we can see Maths has more outliers which is 8 while Readng has 6 and writing has 5.



Outlier may exist due to natural variartion in student performance,where some students score significantly lower or higher than others.
They can also occur due to data entry error or circumstances such as lack of preparation or absence during exam.
